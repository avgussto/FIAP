#Exemplo 1


##$\lim_{x \to 1} \frac{x^2 + x - 2}{x-2}$

```python
#definir o x simbolo
x = sp.Symbol('x')
x
```

```python
#definir a funcao
funcao1 = (x**2 + x - 2) / (x - 1)

funcao1
```

```python
#encontrar o limite
sp.limit(funcao1, x,1)
```

```python
sp.plot(funcao1)
```

#Exemplo 3 e 4

##$\lim_{x\to 0} 2+3x$

##$\lim_{x\to \infty} 2+3x$

```python
x = sp.Symbol('x')
x
```

```python
funcao3 = 2 + 3 * x
funcao3
```

```python
#x tendendo a 0
sp.limit(funcao3, x, 0)
```

```python
#x tendendo a infinito
sp.limit(funcao3,x,sp.oo)
```

```python
sp.plot(funcao3)
```

#Exemplo 5 e 6

##$\lim_{x\to0} \beta_0 + \beta_1 x$
##$\lim_{x\to } \beta_0 + \beta_1 x$

```python
#definir o simbolo beta
x = sp.Symbol ('x')
beta0 = sp.Symbol('beta_0')
beta1 = sp.Symbol('beta_1')
beta0
```

```python
#definir a funcao
funcao4 = beta0 + beta1 * x
funcao4
```

```python
#limite x tendendo a 0
sp.limit(funcao4, x, 0)
```

```python
#limite x tendendo a infinito
sp.limit(funcao4, x, sp.oo)
```

#Exemplo 7 e 8

##$\lim_{x\to0} \frac{1}{1+e^{-x}}$

##$\lim_{x\to} \frac{1}{1+e^{-x}}$

```python
#definir o x
x = sp.Symbol('x')
x
```

```python
#funcao
funcao5 = 1/ (1 + sp.exp(-x))
funcao5
```

```python
#limite tendendo a 0
sp.limit(funcao5,x,0)
```

```python
#limite x tendendo a infinito
sp.limit(funcao5,x,sp.oo)
```

```python
#limite x tendendo a infinito negativo
sp.limit(funcao5,x,-sp.oo)
```

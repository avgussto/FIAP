##Exemplo 3

```python
# definir o simbolo
x = sp.Symbol('x')
x
```

```python
#definir a equacao
funcao2 = (x**2+3) * (2*x+1)
funcao2
```

```python
#eccontrar a derivada
funcao_derivada2 = sp.diff(funcao2)
funcao_derivada2
```

```python
#simplificar a funcao
sp.simplify(funcao_derivada2)
```

* transformar a funcao simbolica em numerica

```python
funcao_derivada_calc= sp.lambdify(x, funcao_derivada2)

funcao_derivada_calc(2)
```

#Exemplo do lucro

```python
#definir o simbolo
p = sp.Symbol('p')
p
```

```python
#definir a equacao
funcao_lucro = -2 * p**2 + 50.4 * p - 10
funcao_lucro
```

```python
#encontrar a derivada
funcao_lucro_derivada = sp.diff(funcao_lucro)
```

```python
#resolvendo a derivada
sp.solve(funcao_lucro_derivada)
```

- grafico do lucro


```python
#tornando a funcao calculavel
funcao_lucro_calc = sp.lambdify(p,funcao_lucro)
```

```python
funcao_lucro_calc(12.60)
```

```python
precos = list(range(1,20))
precos
```

```python
#testando diferentes valores de lucro
lucros = [funcao_lucro_calc(p) for p in precos]
```

```python
#mostrando em grafico os  lucros
plt.plot(precos,lucros)
```

```python
import pandas as pd
import pandas as np
import matplotlib.pyplot as plt
```

# Dataframe com os valores nulos inclusos

```python
df = pd.read_csv('https://raw.githubusercontent.com/diogenesjusto/FIAP/master/dados/classifier.csv')
```

```python
df[['Score1']].describe()
```

```python
a1 = df.iloc[:3000,:]
```

```python
a1[['Score1']].describe()
```

```python
a2 = df.sample(3000,random_state=4)
a2[['Score1']].describe()
```

```python
df['Score1'].hist()
```

```python
a1['Score1'].hist()
```

```python
a2['Score1'].hist()
```

```python
# bibliotecas
from pycaret.classification import *
import pandas as pd
```

```python
# df
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/Gradua%C3%A7%C3%A3o/dados/Fraud_Study_Enrich.csv")
df.head()
```

```python
# Configurando PyCaret usando 'RESULTADO_N' como a coluna-alvo
clf_setup = setup(data=df, target='RESULTADO_N', session_id=123, fix_imbalance=True)
```

```python
# Comparando Modelos
melhor_modelo = compare_models()
```

```python
# Fazendo tuning no modelo escolhido
modelo_ajustado = tune_model(melhor_modelo)
```

```python
# Plotando as features
plot_model(modelo_ajustado, plot='feature')
```

```python
# Avaliando o Modelo
evaluate_model(modelo_ajustado)
```

```python
# Salvando o Modelo
save_model(modelo_ajustado, 'modelo_fraude')
```
